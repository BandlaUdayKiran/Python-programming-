# Mutable And Immutable

 we must be aware that everything in Python is an object. However, there is an important and notable difference between different types of objects in Python. Every types in Python is an object and all objects in python are either mutable or immutable types.

### Mutable DataType

Mutable objects are those that allow you to change their value or data in place without affecting the object’s identity . In simple words mutable data types are those data types which we can be change or modify even after runtime.
Ex: List , Set , Dictionary are mutable data types .

```python
#python code to test that list is mutable.
Numbers = [1,2,3,4,5]
print(Numbers)
print(id(Numbers))
Numbers[3]=10
print(Numbers)
print(id(Numbers))
```

### Immutable DataType

An immutable data type in Python is an object whose state cannot be change or modify after it has been created. In simple words, the value assigned to a variable cannot be changed for the immutable data types. For example, String is an immutable data type in Python. We cannot change its content, otherwise, we may fall into a TypeError. Even if we assign any new content to immutable objects, then a new object is created (instead of the original being modified).

Ex: Numbers , Strings , Tuples are immutable data types .

```python
#python code to test that string is immutable.
a = "Data science "
a[3]='p' 
print(a) #it gives us an error when we change or modify the string because string is immutable datatype..
```

# slicing:
 Slicing is the process of accessing a sub-string or sub-sequence of a string or a sequence by specifying a starting and ending index. In Python, you perform slicing using the colon (:) operator. The syntax for slicing is as follows:

sequence[start_index : end_index : step_size]

where start_index is the index of the first element in the sub-sequence, end_index is the index of the last element in the sub-sequence (excluding the element at the end_index) and step_size by default +1. To slice a sequence, you can use square brackets [ ] with the start and end indices separated by a colon.

```python
name = "my name is uday kiran"
name
```

```python
#slicing:
name[11:21]
```

```python
#adding step size 2
name[2:21:2] 
```

### slice from the start:
By leaving out the start index, the range will start at the first character:

```python
name = "my name is uday kiran"
name[:11]
```

```python
name[:15]
```

```python
name[:21]
```

### slice from the end:
By leaving out the end index, the range will go to the end:

```python
name = "my name is uday kiran"
name
```

```python
name[16:]
```

```python
name[11:]
```

```python
name[3:] 
```

The substring always includes the character at the start and excludes the string at the end. The start and end are optional. If you omit the start, it defaults to zero. If you omit the end, it defaults to the string length. 

```python
name[::]
```

Reversing the string:

```python
name[::-1]
```

### Negitive slicing:

```python
name[-10:-6]
```

```python
name[-21:-1]
```

```python
name[-21:-5:2] #addind step size 2
```

### Repetation of strings

Sometimes we need to repeat the string in the program, and we can do this easily by using the repetition operator in Python.The repetition operator is denoted by a '*' symbol and is useful for repeating strings to a certain length.

```python
str = "hello world"
str
```

```python
str*3
```

Similarly, it is also possible to repeat any part of the string by slicing:

```python
str[1:3]*5
```
