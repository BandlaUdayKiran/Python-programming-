```python
#binary numbers
x = 5
print(x,type(x))

x = bin(x)  #converting int value to binary value. funtion bin()-->binary.
print(x,type(x))
```

```python
#octa decimal numbers
x = 17
print(x,type(x))

x = oct(x)  #converting int value to octa decimal. funtion oct()-->octa decimal.
print(x,type(x))
```

```python
#hexa decimal numbers
x = 17
print(x,type(x))

x = hex(x)  #converting int value to hexa decimal. funtion hex()-->octa decimal.
print(x,type(x))
```

```python
#converting the bin,oct & hex values to int.
print(int('0x11',0) ,int('0o21',0),int('0b101',0))
print(int('0x11',0) ,int('0o21',8),int('0b101',2))
```

```python
# int() with user defined bases
#int() base is must be between >=2 and <=36
print(int('0011',3), int('0011',17))
print((int('0011',7)))
print((int('00110',7)))
```

```python
#slicing
#syntax:[start_index:end_index:step] By default step = 1. 
y = 'classroom rules'
print(y[5:11])
print(y[:11])
print(y[5:])
print(y[5:11:2])
```

```python
#list:list is a heterogeneous container.It can contain int,float,complex,string,boolen.
x = [1,2.2,5+3j,print,'hello',True]
print(x[3])
print(x[:4])
print(x[2:])
print(x[-2])   #negitive indexing in python starts from -1
print(x[-1:-5])
print(x[::-1]) #reversing syntax:[::-1]
```
