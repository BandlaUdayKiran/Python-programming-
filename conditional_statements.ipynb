# Conditional statements:
Conditional statements are an essential part of programming in Python. They allow you to make decisions based on the values of variables or the result of comparisons.

### if statement in python:
The if statement is used to test a particular condition and if condition is true, it execute a block of code known as if block. The condition of if statement can be any valid logical expression which can be either evaluated to true or false.

The if statement checks the condition first. If the condition evaluate to "True", it execute the statement in the if-block otherwise, it ignores the statement.

The colon(:) that follows the condition is very k=important. If you forget it , you'll get a syntax error.

python relies on indentation to define scope in the code. If statement without indentation will raise an error.

##### Syntax:
if condition:

    # code to execute if condition is true

### if else statement:
If the condition is True, the code block indented below the if statement will be executed, and the code block indented below the else statement will be skipped.

If the condition is False, the code block indented below the else statement will be executed, and the code block indented below the if statement will be skipped.

##### Syntax:
if condition:

    # code to execute if condition is true

else:

    # code to execute if condition is false

```python
a, b = int (input('enter a value of a: ')), int(input('enter a value of b: '))

if a>b :
    print(f'{a} is greater than {b}')
else:
    print(f'{a} is less than {b}')
```

### if elif else statement:
The elif statement is used to check for multiple conditions and execute the code block within if any of the conditions evaluate to be true.

The elif statement is similar to the else statement in the context that it is optional but unlike the else statement, there can be multiple elif statements in a code block

##### Syntax:
if condition1:

    # code to execute if condition1 is true

elif condition2:

    # code to execute if condition1 is false and condition2 is true

elif condition3:

    # code to execute if condition1 and condition2 are false, and condition3 is true

else:

    # code to execute if all conditions are false

```python
a = int(input('enter a number: '))
b = int(input('enter a number: '))
if a>b:
    print(f'{a} is greater than b')
elif a<b:
    print(f'{a} is less than b')
else:
    print(f'{a} is equal to {b}')
```

1)write a python code which accepts a three or more digits number and prints :
'A' if the last two digits of the number is divisible by 7.

'B' if the last two digits of number is divisible by 3.

'AB' if the last two digits of the number is divisible by bothe 7 & 3.

'Z' if last two digits of number is neither divisible by 7 nor 3.

```python
num = input('enter a three or greater than three digit number: ')

if len(num) >= 3 and (num.isdigit()):
    
    if (int(num[-2:]) % 3 == 0) and (int(num[-2:]) % 7 == 0):
        print('AB')
    elif int(num[-2:]) % 3 == 0:
        print('B')
    elif int(num[-2:]) % 7 == 0:
        print('A')
    else:
        print('Z')
        
        
else:
    print("try entering a three digit number or greater than three digit positive integer")
```

2)write a code which accepts an alphabet from user and prints 'vowel' if the the alphabet is an vowel, 'not vowel' if the alphabet is not a vowel

```python
x = input('enter an alphabet: ').lower()

if len(x) == 1 and x.isalpha():
    if x in ['a','e','i','o','u']:
        print('vowel')
    else:
        print('not vowel')
        
else:
    print('enter a single alphabet')
```

3)write a code to accepts four sides of an quadrilateral and print if it is a square or rectangle or an irregular quadrilateral.
